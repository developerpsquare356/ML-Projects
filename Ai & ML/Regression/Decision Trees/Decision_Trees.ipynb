```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Training the Decision Tree Regression model on the whole dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

```python
# predicting a new result with Decision Tree Regression
print( regressor.predict([[6.5]]))
```

```python
# visualising the Decision Tree Regression results (for higher resolution and smoother curve)

# x_grid = np.arange(min(X.flatten()), max(X.flatten()), 0.01) # it creates a vector of values from the minimum value of X to the maximum value of X with a step of 0.01
# x_grid = x_grid.reshape(-1, 1) # it reshapes the x_grid vector into a 2D array with one column and as many rows as there are values in x_grid


plt.scatter(X, y, color='red') 
plt.plot(X, regressor.predict(X), color='blue') 
plt.title('Truth or Bluff (Decision Tree Regression)') 
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# r2 Score
from sklearn.metrics import r2_score
print(r2_score(y, regressor.predict(X)))
```
