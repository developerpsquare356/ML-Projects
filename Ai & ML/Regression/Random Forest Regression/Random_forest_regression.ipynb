```python
# including libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing the data sets
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
#  train the random forest regression model on the whole data set
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor.fit(X, y)
```

```python
regressor.predict([[6.5]])
```

```python
# visualising the random forest regression results (for higher resolution and smoother curve)
X_grid = np.arange(min(X.flatten()), max(X.flatten()), 0.01) # it will create a vector of values from the minimum value of X to the maximum value of X with a step of 0.01
X_grid = X_grid.reshape((len(X_grid), 1)) # it will reshape the X_grid vector into a matrix with one column and as many rows as there are values in X_grid
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# r2 Score
from sklearn.metrics import r2_score
print(r2_score(X_grid, regressor.predict(X_grid)))
```
